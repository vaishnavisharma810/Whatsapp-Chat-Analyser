```python
import re  #re stands for regular expression
import pandas as pd
```

```python
f = open('WhatsApp Chat with IOTP-33 PRISM Worklet.txt', 'r', encoding='utf-8')
```

```python
data = f.read()
```

```python
print(data)
```

```python
pattern = '\d{1,2}/\d{1,2}/\d{2,4},\s\d{1,2}:\d{2}\s-\s'
```

```python
messages = re.split(pattern, data)[1:] #[1:] was used to remove a empty string in the starting of the messages variable 
messages #meassages is a list
```

```python
dates = re.findall(pattern, data)
dates
```

```python
df = pd.DataFrame({'user_message':messages, 'message_date':dates})
```

```python
#convert message_date type
df['message_date'] = pd.to_datetime(df['message_date'], format='%m/%d/%y, %H:%M - ')  #y for two digits in year, Y for four digits
```

```python
df.rename(columns={'message_date':'date'}, inplace=True)
df.head()
```

```python
df.shape
```

```python
#to separate user name and message
users = []
messages = []
for message in df['user_message']:
    entry = re.split('([\w\W]+?):\s', message)
    if entry[1:]: #user name
        users.append(entry[1])
        messages.append(entry[2])
    else:
        users.append('group notification')
        messages.append(entry[0])
df['user'] = users
df['messages'] = messages
df.drop(columns = ['user_message'], inplace=True)
df.head()
```

```python
#breaking down date in date, month and year
df['year'] = df['date'].dt.year
```

```python
df.head()
```

```python
df['month'] = df['date'].dt.month_name()
```

```python
df.head()
```

```python
df['day'] = df['date'].dt.day
```

```python
df.head()
```

```python
df['hour'] = df['date'].dt.hour
```

```python
df['minute'] = df['date'].dt.minute
```
